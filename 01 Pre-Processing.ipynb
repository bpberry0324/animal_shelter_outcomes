```python
import pandas as pd
```

```python
intakes = pd.read_csv('datasets/Austin_Animal_Center_Intakes.csv')
outcomes = pd.read_csv('datasets/Austin_Animal_Center_Outcomes.csv')
```

```python
#ceaning column names

outcomes.columns = outcomes.columns.str.lower().str.replace(' ', '_')
intakes.columns = intakes.columns.str.lower().str.replace(' ', '_')
```

```python
intakes.head()
```

```python
outcomes.head()
```

```python
# removing improper data entires
outcomes = outcomes[~(outcomes['age_upon_outcome'] == '0 years')]
outcomes = outcomes[~(outcomes['age_upon_outcome'] == '-1 years')]
outcomes = outcomes[~(outcomes['age_upon_outcome'] == '-2 years')]
outcomes = outcomes[~(outcomes['age_upon_outcome'] == '-3 years')]
outcomes = outcomes[~(outcomes['age_upon_outcome'].isnull())]
```

```python
outcomes.shape
```

```python
def clean_age(age):
    """
        Takes a string form of age, and transforms into an a floating point number.
    """
    age_f = age.split(' ')
    if (age_f[1] == 'day' or age_f[1] == 'days'):
        return round(float(age_f[0])/365, 3)
    elif (age_f[1] == 'week' or age_f[1] == 'weeks'):
        return round(float(age_f[0])/52.143, 3)
    elif (age_f[1] == 'month' or age_f[1] == 'months'):
        return round(float(age_f[0])/12,3)
    else:
        return age[0]
```

```python
outcomes['age_upon_outcome'] = outcomes['age_upon_outcome'].apply(clean_age)
intakes['age_upon_intake'] = intakes['age_upon_intake'].apply(clean_age)
```

```python
outcomes['is_named'] = outcomes['name'].notnull().astype(int)
```

```python
outcomes.head()
```

```python
# Dropping duplicate column
outcomes.drop(columns=['monthyear'], axis=1, inplace = True)
```

```python
outcomes['datetime'] = pd.to_dateframe[outcomes['datetime']]
```

```python
outcomes['weekday'] = pd.to_datetime(outcomes['datetime']).apply(lambda x: x.day_name())
outcomes['month'] = pd.to_datetime(outcomes['datetime']).apply(lambda x: x.month_name())
outcomes['year'] = pd.to_datetime(outcomes['datetime']).apply(lambda x: x.year())
```

```python
outcomes['year'] = pd.to_datetime(outcomes['datetime']).apply(lambda x: x.qyear())
```
